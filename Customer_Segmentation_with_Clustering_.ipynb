```python
# Function to detect outliers using the IQR method
# Note this is just a quick code to detect outliers on numeric columns and will include label features such as Order ID
def detect_outliers_iqr(df):
    outliers = {}
    for col in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers
```

```python
# Detect outliers using IQR method
iqr_outliers = detect_outliers_iqr(sample_df)
print("\nOutliers detected using the IQR method:")
for col, out in iqr_outliers.items():
    print(f"{col}: {len(out)} outliers")
```

# 2. Feature Engineering

```python
# Create new feature Frequency
sample_df['Frequency' ] = sample_df.groupby( 'Customer ID' )['Order ID' ].transform( 'count')

# Specify the format for Delivery_Date
sample_df['Delivery_Date'] = pd.to_datetime(sample_df['Delivery_Date'], format='%d%b%Y')

# Create Recency (time since last order, shorter time/lower recency is good)
sample_df['Recency'] = (pd.to_datetime('today') - sample_df['Delivery_Date']).dt.days

# Define a function to convert currency to float
def currency_to_float(currency_str):
    if isinstance(currency_str, str):
        return float(currency_str.replace('$', '').replace(',', ''))
    return currency_str

# Create Customer Lifetime Value (CLV)
sample_df['Total Revenue'] = sample_df['Total Revenue'].apply(currency_to_float)

# Create Average Unit Cost
sample_df['Unit Cost'] = sample_df['Unit Cost'].apply(currency_to_float)

# Convert the DOB to datetime
sample_df[ 'DOB'] = pd.to_datetime(sample_df[ 'Customer_BirthDate' ], format= '%d%b%Y')

# Calculate age
today = pd.Timestamp( 'now')
sample_df[ 'Age'] = (today - sample_df[ 'DOB']).dt.days // 365

# Drop the columns that won't be used in clustering
df_for_clustering = sample_df[['Customer ID', 'Frequency', 'Recency', 'Total Revenue', 'Unit Cost', 'Age']]

# Exclude 'Customer ID' from the clustering features
clustering_features = df_for_clustering.drop(columns=['Customer ID'])

# Numerical feature list
numerical_features = clustering_features.columns.tolist()
```

#3. Clustering with Machine Learning Models

```python
# Numerical feature list
numerical_features = clustering_features.columns.tolist()

# Pipeline for preprocessing
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numerical_features)])
```

## a. K-means Clustering

```python
# Function to calculate WCSS and silhouette scores
def calculate_wcss_silhouette(df, max_iter=300, k_range=range(2, 11)):
    wcss = []
    silhouette_scores = []

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto', max_iter=max_iter)
        pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('clusterer', kmeans)])
        pipeline.fit(df)
        wcss.append(pipeline['clusterer'].inertia_)  # Inertia is the WCSS
        silhouette_scores.append(silhouette_score(df, pipeline['clusterer'].labels_))

    return k_range, wcss, silhouette_scores
```

```python
# Function to plot WCSS (Elbow Method)
def plot_elbow_method(k_range, wcss):
    plt.figure(figsize=(10, 8))
    plt.plot(k_range, wcss, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.title('Elbow Method for Optimal Number of Clusters')
    plt.show()
```

```python
# Function to plot silhouette scores
def plot_silhouette_scores(k_range, silhouette_scores):
    plt.figure(figsize=(10, 8))
    plt.plot(k_range, silhouette_scores, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Scores for Optimal Number of Clusters')
    plt.show()
```

```python
# Run the calculation function
k_range, wcss, silhouette_scores = calculate_wcss_silhouette(clustering_features, max_iter=300, k_range=range(2, 11))
```

```python
# Plot the results
plot_elbow_method(k_range, wcss)
plot_silhouette_scores(k_range, silhouette_scores)
```

Elbow Method for Optimal Number of Clusters: The "elbow" in the plot, which indicates the point where adding more clusters does not significantly decrease the WCSS (within-cluster sum of squares), seems to occur at around 4 or 5 clusters.

Silhouette Scores for Optimal Number of Clusters: the silhouette score is highest for 2 clusters, indicating that 2 clusters may be the optimal choice. However, this score sharply declines as the number of clusters increases, suggesting that increasing the number of clusters reduces the quality of the clustering.

Hierarchical clustering will be carried out to visualise the plots to better inform optimal number of clusters...




## b. Hierarchical Clustering

```python
# Function to perform hierarchical clustering using scikit-learn
def hierarchical_clustering_sklearn(data, n_clusters=5, linkage_method='average'):
    # Perform Agglomerative Clustering
    agglo_cluster = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_method, metric='euclidean')
    cluster_labels = agglo_cluster.fit_predict(data)
    return cluster_labels

# Creating the linkage matrix and plotting the dendrogram using SciPy
def plot_dendrogram(data, method='average'):
    # Create the linkage matrix
    Z = linkage(data, method=method)

    # Plot the dendrogram
    plt.figure(figsize=(15, 10))
    dendrogram(Z)
    plt.title('Dendrogram')
    plt.xlabel('Sample Index or (Cluster Size)')
    plt.ylabel('Distance')
    plt.show()

# Perform hierarchical clustering with average linkage for 5 clusters
sklearn_clusters_avg_5 = hierarchical_clustering_sklearn(clustering_features, n_clusters=5, linkage_method='average')
sample_df['sklearn_cluster_avg_5'] = sklearn_clusters_avg_5

# Perform hierarchical clustering with average linkage for 2 clusters
sklearn_clusters_avg_2 = hierarchical_clustering_sklearn(clustering_features, n_clusters=2, linkage_method='average')
sample_df['sklearn_cluster_avg_2'] = sklearn_clusters_avg_2

# Perform hierarchical clustering with Ward linkage for 5 clusters
sklearn_clusters_ward_5 = hierarchical_clustering_sklearn(clustering_features, n_clusters=5, linkage_method='ward')
sample_df['sklearn_cluster_ward_5'] = sklearn_clusters_ward_5

# Perform hierarchical clustering with Ward linkage for 2 clusters
sklearn_clusters_ward_2 = hierarchical_clustering_sklearn(clustering_features, n_clusters=2, linkage_method='ward')
sample_df['sklearn_cluster_ward_2'] = sklearn_clusters_ward_2

# Plot the dendrogram for average linkage
plot_dendrogram(clustering_features, method='average')

# Plot the dendrogram for Ward linkage
plot_dendrogram(clustering_features, method='ward')

# Function to count and print the number of customers in each cluster
def print_cluster_counts(cluster_labels, title):
    cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
    print(f"Customer counts in each cluster for {title}:")
    print(cluster_counts)
    print()

# Print the counts for each clustering solution
print_cluster_counts(sklearn_clusters_avg_5, 'Average Linkage with 5 Clusters')
print_cluster_counts(sklearn_clusters_avg_2, 'Average Linkage with 2 Clusters')
print_cluster_counts(sklearn_clusters_ward_5, 'Ward Linkage with 5 Clusters')
print_cluster_counts(sklearn_clusters_ward_2, 'Ward Linkage with 2 Clusters')
```

```python
# Calculate silhouette scores for the existing clusters
silhouette_avg_5 = silhouette_score(clustering_features, sample_df['sklearn_cluster_avg_5'])
silhouette_avg_2 = silhouette_score(clustering_features, sample_df['sklearn_cluster_avg_2'])
silhouette_ward_5 = silhouette_score(clustering_features, sample_df['sklearn_cluster_ward_5'])
silhouette_ward_2 = silhouette_score(clustering_features, sample_df['sklearn_cluster_ward_2'])

# Print the silhouette scores
print(f"Silhouette Score for Average Linkage with 5 Clusters: {silhouette_avg_5}")
print(f"Silhouette Score for Average Linkage with 2 Clusters: {silhouette_avg_2}")
print(f"Silhouette Score for Ward Linkage with 5 Clusters: {silhouette_ward_5}")
print(f"Silhouette Score for Ward Linkage with 2 Clusters: {silhouette_ward_2}")
```

```python
# Print the counts for each clustering solution
print_cluster_counts(sklearn_clusters_avg_5, 'Average Linkage with 5 Clusters')
print_cluster_counts(sklearn_clusters_avg_2, 'Average Linkage with 2 Clusters')
print_cluster_counts(sklearn_clusters_ward_5, 'Ward Linkage with 5 Clusters')
print_cluster_counts(sklearn_clusters_ward_2, 'Ward Linkage with 2 Clusters')
```

# 5. Dimensionality Reduction

```python
# Existing PCA and t-SNE code
def apply_pca(data, n_components=2):
    pca = PCA(n_components=n_components)
    reduced_data = pca.fit_transform(data)
    return pca, reduced_data

# Function to apply t-SNE on PCA-reduced data and reduce to 2 dimensions
def apply_tsne(data, n_components=2, perplexity=30, n_iter=1000):
    tsne = TSNE(n_components=n_components, perplexity=perplexity, n_iter=n_iter, random_state=42)
    reduced_data = tsne.fit_transform(data)
    return reduced_data

# Fit PCA to the data and get the PCA object and reduced data
pca, pca_data = apply_pca(clustering_features)

# Apply t-SNE on PCA-reduced data
tsne_data = apply_tsne(pca_data)

# Add the PCA and t-SNE results to the dataframe for plotting
df_for_clustering['PCA1'] = pca_data[:, 0]
df_for_clustering['PCA2'] = pca_data[:, 1]
df_for_clustering['tSNE1'] = tsne_data[:, 0]
df_for_clustering['tSNE2'] = tsne_data[:, 1]

# Plot PCA
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='viridis', data=df_for_clustering)
plt.title('PCA of Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Plot t-SNE
plt.subplot(1, 2, 2)
sns.scatterplot(x='tSNE1', y='tSNE2', hue='Cluster', palette='viridis', data=df_for_clustering)
plt.title('t-SNE of PCA-Reduced Clusters')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')

plt.tight_layout()
plt.show()

# Obtain the explained variance ratio from the PCA object
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the explained variance ratio
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(np.cumsum(explained_variance_ratio)) + 1), np.cumsum(explained_variance_ratio), where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.title('PCA Explained Variance Ratio')
plt.legend(loc='best')
plt.show()
```

### **Thoughts and observations**

#### PCA Component 1 and Component 2
- The PCA plot shows a clear separation of clusters along the two principal components.
- This indicates that the clusters have distinct characteristics that can be captured by linear combinations of the original features.

#### t-SNE Component 1 and Component 2
- The t-SNE plot shows a non-linear projection of the PCA-reduced data.
- This provides a more nuanced view of the cluster separations, indicating that the clusters are well-defined and distinct.

### Explained Variance

#### Individual Explained Variance
- The first principal component (PCA Component 1) explains a significant portion of the variance in the data, approximately 85%.
- The second principal component (PCA Component 2) explains a smaller portion of the variance, around 15%.

#### Cumulative Explained Variance
- The cumulative explained variance reaches nearly 100% with the first two components combined.
- This indicates that the two principal components capture almost all the variance in the data.
